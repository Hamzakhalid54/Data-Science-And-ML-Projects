```python
import pandas as pd
file_path = 'exercises_problem2 (1).parquet'
df = pd.read_parquet(file_path)
```

```python
df.tail()
```

```python
df.info()
```

```python
import matplotlib.pyplot as plt
import pandas as pd

plt.figure(figsize=(8, 6))
for label in df['class'].unique():
    subset = df[df['class'] == label]
    plt.scatter(subset['var1'], subset['var2'], label=f"Class {label}", alpha=0.7)
plt.xlabel('var1',fontsize=14)
plt.ylabel('var2',fontsize=14)
plt.legend()
plt.title('Scatter Plot of var1 vs var2 by Class')
plt.show()
```
