```python
import matplotlib.pyplot as plt
import numpy as np

# Define the points
x = np.array([3, 7])
y = np.array([5, 2])

# Calculate the coefficients of the line
coefficients = np.polyfit(x, y, 1)

# Create a polynomial from the coefficients (this represents our line)
polynomial = np.poly1d(coefficients)

# Generate x values for the line to cover wider range than just the points
x_line = np.linspace(0, 10, 100)

# Calculate y values for the line
y_line = polynomial(x_line)

# Create the plot
plt.figure(figsize=(8, 8))

# Plot the points
plt.plot(x, y, 'ko', markersize=8, label='Points')  # 'ko' for black circles

# Plot the line
plt.plot(x_line, y_line, 'k-', linewidth=2, label='Line through points')  # 'k-' for a black line

# Set the range for axes
plt.xlim(0, 10)
plt.ylim(0, 6)

# Label axes
plt.xlabel('X-axis', fontsize=14)
plt.ylabel('Y-axis', fontsize=14)

# Set the font size for tick marks
plt.xticks(fontsize=14)
plt.yticks(fontsize=14)

# Add a grid
plt.grid(True, which='both', linestyle='--', linewidth=0.5)

# Add a legend
plt.legend(fontsize=14)

# Show plot
plt.show()
```
