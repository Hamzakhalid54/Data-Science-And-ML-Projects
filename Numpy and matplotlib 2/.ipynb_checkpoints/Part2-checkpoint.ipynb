```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
file_path = 'exercises_problem2 (1).parquet'
df = pd.read_parquet(file_path)
```

```python
fig, axes = plt.subplots(1, 3, figsize=(18, 6))
variables = ['var1', 'var2', 'var3']
for ax, var in zip(axes, variables):
    df.boxplot(column=var, by='class', ax=ax)
    ax.set_title(f'Boxplot of {var} by Class')
    ax.set_xlabel('Class')
    ax.set_ylabel(var)
plt.suptitle('')
plt.show()
```

```python
for var in variables:
    plt.figure(figsize=(8, 6))
    for label in df['class'].unique():
        subset = df[df['class'] == label]
        plt.hist(subset[var], alpha=0.7, label=f"Class {label}")
    plt.xlabel(var, fontsize=14)
    plt.ylabel('Frequency', fontsize=14)
    plt.legend()
    plt.title(f'Histogram of {var} by Class')
    plt.show()
```

```python
class_counts = df['class'].value_counts().sort_index()
class_counts.plot(kind='bar', stacked=True, figsize=(8, 6))
plt.title('Class Counts')
plt.xlabel('Class')
plt.ylabel('Count')
plt.show()
```

```python
for var in variables:
    plt.figure(figsize=(8, 6))
    for label in df['class'].unique():
        subset = df[df['class'] == label][var]
        plt.hist(subset, bins=30, alpha=0.5, label=f"Class {label}", density=True)
    plt.xlabel(var, fontsize=14)
    plt.ylabel('Density', fontsize=14)
    plt.legend()
    plt.title(f'Density Plot of {var} by Class')
    plt.show()
```

```python
plt.figure(figsize=(8, 6))
plt.matshow(df.corr(), fignum=1)
plt.colorbar()
plt.xticks(range(len(df.corr().columns)), df.corr().columns, rotation=90)
plt.yticks(range(len(df.corr().columns)), df.corr().columns)
plt.title('Correlation Heatmap', pad=20)
plt.show()
```
