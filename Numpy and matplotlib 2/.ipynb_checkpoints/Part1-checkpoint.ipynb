```python
import numpy as np
import matplotlib.pyplot as plt
data_x = np.array([3, 7])
data_y = np.array([5, 2])


line_fit = np.polyfit(data_x, data_y, 1)

# Polynomial representation of the fit line
fit_poly = np.poly1d(line_fit)

# Generate x values for plotting the line
plot_x = np.linspace(0, 10, 100)

# Generate y values using the polynomial
plot_y = fit_poly(plot_x)

# Set up the plot with figure size
fig, ax = plt.subplots(figsize=(8, 8))

# Plot original data points
ax.scatter(data_x, data_y, color='black', s=50, label='Data Points')

# Plot the fit line
ax.plot(plot_x, plot_y, color='black', linestyle='-', linewidth=2, label='Fit Line')

# Setting axes limits
ax.set_xlim([0, 10])
ax.set_ylim([0, 6])

# Axes labels
ax.set_xlabel('X-axis', fontsize=14)
ax.set_ylabel('Y-axis', fontsize=14)

# Tick parameters
ax.tick_params(axis='both', which='major', labelsize=14)

# Grid
ax.grid(True, which='both', linestyle='--', linewidth=0.5)

# Legend
ax.legend(fontsize=14)

# Display the plot
plt.show()
```
