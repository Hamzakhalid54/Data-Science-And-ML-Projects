```python
import matplotlib.pyplot as plt
import numpy as np

# Points definition
points_x = np.array([3, 7])
points_y = np.array([5, 2])

# Fitting a line to the points
fit_coeffs = np.polyfit(points_x, points_y, 1)

# Function for the fitted line
def fitted_line(x):
    return fit_coeffs[0] * x + fit_coeffs[1]

# Generating x and y values for the fitted line
line_x = np.linspace(0, 10, 100)
line_y = fitted_line(line_x)

# Plotting starts here
fig = plt.figure(figsize=(8, 8))
graph_area = fig.add_subplot(111)

# Plotting the data points
graph_area.plot(points_x, points_y, 'o', color='black', markersize=8, label='Points of data')

# Plotting the fitted line
graph_area.plot(line_x, line_y, color='gray', linestyle='-', linewidth=2, label='Line Fitted')

# Setting the visual appearance of the plot
graph_area.set_xlim(0, 10)
graph_area.set_ylim(0, 6)
graph_area.set_xlabel('X-axis', fontsize=14)
graph_area.set_ylabel('Y-axis', fontsize=14)
graph_area.tick_params(axis='both', labelsize=14)
graph_area.grid(which='both', linestyle='--', linewidth=0.5)
graph_area.legend(fontsize=14)

plt.show()
```
